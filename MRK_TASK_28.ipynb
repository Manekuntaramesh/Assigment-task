```python
import numpy as np
import pandas as pd
```

```python
RK['Position'] = np.where(RK['Position'].isna(),RK['Position'].mode()[0],RK['Position'])
```

```python
RK['State'] = np.where(RK['State'].isna(),RK['State'].mode()[0],RK['State'])
```

```python
RK["Zip"].fillna(RK['Zip'].mode()[0], inplace = True)
```

```python
RK["Gender"].fillna(RK['Gender'].mode()[0], inplace = True)
```

```python
RK["MaritalDesc"].fillna(method = "bfill", inplace = True)
```

```python
RK["CitizenDesc"].fillna(method = "ffill", inplace = True)
```

```python
RK['Department'].fillna(RK['Department'].mode()[0],inplace=True)
```

```python
RK['Salaries'].fillna(RK['Salaries'].mean(),inplace=True)
```

```python
RK["Race"].fillna("White", inplace = True)
```

```python
RK.isna().sum()
```

```python
RK.pivot_table(values='Salaries',index='Position',aggfunc='mean')
```
