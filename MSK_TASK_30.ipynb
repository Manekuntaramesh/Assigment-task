## Data Description

Clean the data..(missing values,outliers,type casting,duplicate,variance,...etc)

```python
import pandas as pd
import numpy as np
```

```python
aa=pd.read_excel(r"C:\Users\manne\Downloads\shopping.xlsx")
```

```python
aa
```

```python
aa.head()
```

```python
aa.drop('Customer ID',axis = 1 , inplace = True)
aa
```

```python
aa.isna().sum()
```

```python
aa['Age'].fillna(aa['Age'].median(),inplace =True)
aa['Gender'].fillna(aa['Gender'].mode()[0],inplace =True)
aa['Item Purchased'].fillna(aa['Item Purchased'].mode()[0],inplace =True)
aa['Category'].fillna(aa['Category'].mode()[0],inplace =True)
aa['Location'].fillna(aa['Location'].mode()[0],inplace =True)
aa['Color'].fillna(aa['Color'].mode()[0],inplace =True)
aa['Season'].fillna(aa['Season'].mode()[0],inplace =True)
aa['Review Rating'].fillna(aa['Review Rating'].mode()[0],inplace =True)
aa['Subscription Status'].fillna(method= 'bfill',inplace =True)
aa['Discount Applied'].fillna(method='ffill',inplace =True)
aa['Promo Code Used'].fillna(aa['Promo Code Used'].mode()[0],inplace =True)
aa['Payment Method'].fillna(method='ffill',inplace =True)
aa['Frequency of Purchases'].fillna(method='ffill',inplace =True)
```

```python
aa.isna().sum()                        # no missing value
```

```python
aa['Age']= np.where(((aa['Age']<0) | (aa['Age']>100)),40,aa['Age'])                  # outliner
aa['Age']
```

```python
aa['Review Rating'] = np.where(((aa['Review Rating']<0)|(aa['Review Rating']>5)),3,aa['Review Rating'])
aa['Review Rating']
```

```python
aa[(aa['Review Rating']<0)|(aa['Review Rating']>5)]
```

```python
aa[aa.duplicated()]       #duplicated
```

```python
aa.duplicated().sum()
```

```python
aa.drop_duplicates(inplace=True)
```

```python
aa.duplicated().sum()
```

```python
aa.info()       ##Type casting
```

```python
aa['Age']=aa['Age'].astype('float16')
aa['Purchase Amount (USD)']=aa['Purchase Amount (USD)'].astype('int8')
aa['Review Rating']=aa['Review Rating'].astype('float16')
aa['Previous Purchases']=aa['Previous Purchases'].astype('int8')
```

```python
def normalization(x):
    n = (x-aa["Age"].min())/(aa["Age"].max()-aa["Age"].min())
    return n
```
